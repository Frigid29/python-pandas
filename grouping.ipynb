```python
#to get the dataframe for the death overs
death_over=df[df["over"]>=16]
```

```python
#to fetch the names of the top 5 batsman who scored most runs in death over
death_over.groupby("batsman")["batsman_runs"].sum().sort_values(ascending=False).head(5)
```

```python
#no of balls played by each player in death_overs
all_batsman=death_over.groupby("batsman")["batsman_runs"].count()
```

```python
#to store the details of all the players who played more than 200 balls in the death overs
x=all_batsman>200
all_player=all_batsman[x].index.tolist()
```

```python
#to store the data of the qualified players
final=death_over[death_over["batsman"].isin(all_player)]
```

```python
#store the runs of players scored in the death over
runs=final.groupby("batsman")["batsman_runs"].sum()
```

```python
#to store the no of balls played in the death over by the players
balls=final.groupby("batsman")["batsman_runs"].count()
```

```python
#to calculate the strike rate
strike_rate=(runs/balls)*100
```
