```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x=np.random.randint(0,50,7)
y=np.random.randint(0,50,7)
np.random.seed(89)
print("x : ",x)
print("y : ",y)
plt.plot(x , y, "*:r", mec='k', mfc="blue", lw=2)
plt.xlabel("x axis values", color='k')
plt.ylabel('y axis values', color='brown')
plt.title("Line Graph", color='c', size=16)
plt.show()
```

```python
print("x : ",x)
print("y : ",y)
colors=np.random.randint(0,100,7)
sizes=10*np.random.randint(50,100,7)
plt.scatter(x,y,c=colors, s=sizes, cmap="Accent", alpha=0.45)
plt.xlabel("X axis")
plt.ylabel("y axis")
plt.title("Scatter Graph")
plt.colorbar()
plt.show()
```

```python
#4 Subplot
a=np.random.randint(0,100,50)
b=np.random.randint(0,100,50)
x=np.random.randint(0,100,5)
y=np.random.randint(0,100,5)
colors=np.random.randint(0,100,50)
sizes=10*np.random.randint(0,100,50)
plt.subplot(2,3,1)
plt.plot(x,y,"o-.b")
plt.xlabel("X axis")
plt.ylabel("y axis")
plt.title("Line Graph", color='c', size=16)

plt.subplot(2,3,2)
plt.scatter(a,b,c=colors, s=sizes, cmap="Accent", alpha=0.45)
plt.xlabel("X axis")
plt.ylabel("y axis")
plt.title("Scatter Graph", color='c', size=16)

plt.subplot(2,3,3)
plt.scatter(a,b,c=colors, s=sizes, cmap="nipy_spectral", alpha=0.45)
plt.xlabel("X axis")
plt.ylabel("y axis")
plt.title("Scatter Graph", color='c', size=16)

plt.subplot(2,3,4)
d=np.array([25,32,29,14])
mylabels=['A', 'B', 'C', 'D']
plt.pie(d, labels=mylabels)
plt.title("Pie Chart", size=16)

plt.subplot(2,3,5)
plt.bar(x,y, color=["red","green","blue","cyan","black"], width=10)
plt.xlabel("X axis")
plt.ylabel("y axis")
plt.title("Bar Graph", size=16)

plt.subplot(2,3,6)
plt.barh(x,y, color=["red","green","blue","cyan","black"], height=10)
plt.xlabel("X axis")
plt.ylabel("y axis")
plt.title("Horizontal Bar Graph", size=16)

plt.subplots_adjust(top=1.2,wspace=0.9, hspace=0.7)
plt.suptitle("super title ",size=20, color="green" ,x=0.5, y=1.341)
plt.show()
```
