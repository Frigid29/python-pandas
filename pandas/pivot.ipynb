```python
import pandas as pd
```

```python
df=pd.DataFrame({"Days":[1,2,3,4,5,6],"eng":[11,16,14,12,13,18],"maths":[15,14,16,12,20,19]})
```

```python
df
```

```python
pd.melt(df)
```

```python
pd.melt(df, id_vars=["Days"])
```

```python
pd.melt(df, id_vars=["Days"],var_name="subject")
```

```python
pd.melt(df, id_vars=["Days"],value_name="marks")
```

```python
pd.melt(df, id_vars=["Days"],value_name="marks",var_name="subject")
```

```python
df2=pd.DataFrame({"Days":[1,2,3,4,5,6],
                  "name":["a","b","a","b","a","b"],
                  "eng":[11,16,14,12,13,18],
                  "maths":[15,14,16,12,20,19]})
df2
```

```python
df2.groupby("name")
```

```python
df2.pivot(index="Days",columns="name")
```

```python
df2.pivot(index="Days",columns="name",values="eng")
```

```python
df3=pd.DataFrame({"Days":[1,1,1,1,2,2],
                  "name":["a","b","a","b","a","b"],
                  "eng":[11,16,14,12,13,18],
                  "maths":[15,14,16,12,20,19]})
df3
```

```python
df3.pivot_table(index="name",columns="Days",aggfunc="mean")
```

```python
df3.pivot_table(index="name",columns="Days",aggfunc="sum",margins=True)
```
