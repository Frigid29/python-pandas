```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.array([2,4,8,10])
```

```python
plt.plot(x)
```

```python
y=np.array([1,9,3,6])
```

```python
plt.plot(y)
```

```python
plt.plot(x,y)
```

```python
#to remove [<matplotlib.lines.Line2D at 0x28fa87af6b0>] we use plt.show
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,marker="*")
```

```python
#ms= marker size 
#its taken in points, where 1point= 1/72 inch
plt.plot(x,y,marker="*",ms=32)
```

```python
#mec= marker edge color, mfc= marker face color
plt.plot(x,y,marker="*",mec='r',mfc='k',ms=20)
```

types of edges

```python
# D for diamond
plt.plot(x,marker="D")
```

```python
# small o 
plt.plot(x,y,marker="o")
```

```python
# lowercase x - to get cross (x) as pointer 
plt.plot(x,y,marker="x")
```

```python
#uppercase X to get a thick cross as marker
plt.plot(x,y,marker="X")
```

```python
# dot(.) to get dot as pointer
plt.plot(x,y,marker=".",ms=20)
```

```python
plt.plot(x,y,marker='2')
```

Lines- there are 4 types of line  
1) "-" solid line 
2) ":" dotted line 
3) "-." dashed/dotted line 
4) "--" dashed line

5) none

```python
#ls stands for line style
plt.plot(x,y,ls="-")
```

```python
#dashed line
plt.plot(x,y,ls="--")
```

```python
#dotted line
plt.plot(x,y,ls=":")
```

```python
#dashed/dotted line
plt.plot(x,y,ls="-.")
```

```python
#no line
plt.plot(x,y,ls="none")
```

```python
#no line
plt.plot(x,y,ls="none",marker="s")
```

color 
r: red; g: green: b: blue; c: cyan; m: magenta; y: yellow; k: black; w: white; 

```python
plt.plot(x,y,color='r')
```

```python
plt.plot(x,y,color='k')
```

```python
plt.plot(x,y,color='c', marker='d',ms=10,mfc='w',mec='k')
```

line width --> lw

```python
plt.plot(y,lw=22,color="c")
```
