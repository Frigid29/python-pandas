```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,9,2,7])
y=np.array([6,3,8,2])
```

```python
font1={'family':'serif', 'color': 'cyan', 'size':20}
```

```python
font2={'family':'serif', 'color': 'blue', 'size':12}
```

```python
font3={'family':'cursive', 'color':'green', 'size': 14}
```

```python
plt.plot(x)
plt.title('fonts example',fontdict=font1)
plt.show()
```

```python
plt.plot(x,y, marker='*', mec='k',mfc='r',ms=16,color='k',ls='-.')
plt.title('fonts example',fontdict=font1)
plt.xlabel("x axis",fontdict=font2)
plt.ylabel("y axis",fontdict=font3)
plt.show()
```
