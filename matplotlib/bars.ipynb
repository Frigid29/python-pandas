```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=['apples', 'bananas']
y=[400, 350]
```

```python
plt.bar(x,y)
```

```python
plt.barh(x,y)
```

```python
colors=['g', 'k']
plt.bar(x,y,color=colors)
```

```python
plt.bar(x,y, width=0.5)
```

```python
plt.barh(x,y, color=colors, height=0.3)
```
