```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x= np.random.randint(0,100,100)
y= np.random.randint(0,100,100)
a= np.random.randint(0,100,100)
b= np.random.randint(0,100,100)
p= np.random.randint(0,100,100)
q= np.random.randint(0,100,100)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y, marker='*')
```

```python
plt.scatter(x,y)
plt.scatter(y,x)
plt.show()
```

```python
colors=np.random.randint(0,100,100)
plt.scatter(x,y,c=colors, cmap='Accent')
plt.show()
```

```python
sizes= 10*np.random.randint(0,100,100)
plt.scatter(x,y, s=sizes)
plt.show()
```

```python
plt.scatter(x,y,c=colors, s=sizes, cmap='nipy_spectral', alpha=0.3)
plt.colorbar()
plt.show()
```

```python
plt.subplot(1,3,1)
plt.scatter(x,y,c=colors, s=sizes, cmap='nipy_spectral', alpha=0.3)
plt.colorbar()

plt.subplot(1,3,2)
plt.scatter(a,b,c=colors, s=sizes, cmap='Accent', alpha=0.3)
plt.colorbar()

plt.subplot(1,3,3)
plt.scatter(p,q,c=colors, s=sizes, cmap='nipy_spectral', alpha=0.3)
plt.colorbar()

plt.subplots_adjust(hspace=1.2, wspace=1.1)
```

```python
fig,axes=plt.subplots(1, 3)
axes[0].scatter(a,b,c=colors, s=sizes, cmap='Accent', alpha=0.3)

axes[1].plot(x)

axes[2].scatter(x,y,c=colors, s=sizes, cmap='Accent', alpha=0.3)

plt.subplots_adjust(wspace=1.1)
```
