```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x=np.random.randint(2,22,5)
print(x)
```

```python
y=np.random.randint(3,33,5)
print(y)
```

```python
np.random.seed(4)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,marker='*', ms=10, mfc='k', mec='hotpink')
plt.title('Both X and Y ')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.subplot(2,2,2)
plt.plot(x, color='green')
plt.plot(y, color='red')
plt.title('both x and y but with different lines')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.subplot(2,2,3)
plt.plot(y)
plt.title('only Y')
plt.xlabel('default values axis')
plt.ylabel('Y')

plt.subplot(2,2,4)
plt.plot(x)
plt.title('only X')
plt.xlabel('default values axis')
plt.ylabel('Y')

plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.show()
```
