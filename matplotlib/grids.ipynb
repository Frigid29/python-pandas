```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x=np.array([2,9,5,1,7])
```

```python
y=np.array([8,6,54,2,21])
```

```python
plt.plot(x)
plt.grid()
```

```python
plt.plot(x)
plt.grid(axis="x")
```

```python
plt.plot(x)
plt.grid(axis="y")
```

```python
plt.plot(x)
plt.grid(ls='-.', color='k', lw=0.5)
```

```python
plt.plot(x, y,color='red', ls='--', lw=1)
plt.grid(ls=':', lw=0.8, color='green')
```
