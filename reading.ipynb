```python
#to fetch a row
df.iloc[0]
```

```python
#iloc is very flexible we can do slicing in it. that is we can fetch multiple rows from it.
df.iloc[5:9]
```

```python
#to fetch data of odd rows between row 1 and row 11
df.iloc[1:11:2]
```

```python
#to fetch data of particular columns and rows
#data.iloc[starting-row-index:ending-row-index,[col1_index, col2.index, col3.index]]
#indexing starts from 0
df.iloc[1:10,[4,5,10]]
```

```python
#to get all rows and some particular columns
#df.iloc[:,[column indexes]]
df.iloc[:,[4,5,10]]
```

```python
#to get multiple col in sequence 
#data.iloc[row-start-index : end-row-index , col-start-index : col-end-index]
df.iloc[20:50,4:10]
```
